```python
# Use 'Species' column for label encoding
le = preprocessing.LabelEncoder()
le.fit(list(data['Species'].unique()))
data['Numeric_species'] = le.transform(list(data['Species'].values))

X1 = data.iloc[:, data.columns != 'Species']
X = X1.iloc[:, X1.columns != 'Numeric_species']
Y = data['Numeric_species']

# Initialize weights based on the number of features
B = np.zeros(X.shape[1])
alpha = 0.1

```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def net_input(B, x):
    return np.dot(x, B)

def cost_function(h, y):
    epsilon = 1e-15  # A small constant to avoid log(0)
    h = np.clip(h, epsilon, 1 - epsilon)  # Clip predicted probabilities to avoid extremes
    return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()

m = len(Y)
epochs = 100


```

```python
for iteration in range(epochs):
    z = net_input(B, X)
    h = sigmoid(z)
    gradient = np.dot(X.T, (h - Y)) / m
    B1 = alpha * gradient
    B = B - B1
    cost = cost_function(h, Y)
    print (B,cost)
```
